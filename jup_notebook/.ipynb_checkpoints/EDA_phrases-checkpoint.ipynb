```python
class ParseTreeNode:
    def __init__(self,val):
        self.value = val
        self.children = []
        self.height = 0
        self.parent = None
        #attributes for generating S and hierarchical positional embeddings
        self.leaf_order_idx = -1 #0 indexed
        self.leaf_list = []
        
    def add_child(self,child):
        self.children.append(child)
        child.parent = self

def printTree(tree,tabs=1):
    print('\t'*tabs,tree.value,tree.height)
    if tree.parent is not None:
        print(tree.parent.value)
    for ch in tree.children:
        printTree(ch,tabs+1)
```

```python
import pickle
data_base_path = '../Data/parallel_data'
# data_base_path 

def read_data(file):
    tree = None
    command = []
    with open(data_base_path + '/trees/' + file + '.tree.pickle','rb') as f:
        tree = pickle.load(f)
    with open(data_base_path + '/' + file + ".cmd.raw",'r') as f:
        for line in f:
            command.append(line.strip())
    return tree, command
```

```python
X,Y = [],[]
x,y = read_data('train.loss_fn0')
X = X+x
Y = Y+y

x,y = read_data('test.loss_fn0')
X = X+x
Y = Y+y

x,y = read_data('valid.loss_fn0')
X = X+x
Y = Y+y
```

```python
def concat_all_words(node):
    if node.height == 0:
        return [node.value]
    word_list = []
    for child in node.children:
        word_list += concat_all_words(child)
    return word_list

def get_phrases(node,height=2):
    if node.height <= height:
        z = concat_all_words(node)
        return [z]
    phrase_list = []
    for child in node.children:
        phrase_list += get_phrases(child,height)
    return phrase_list
    
```

```python
from collections import Counter
phrase_counter = Counter()
for x,y in zip(X,Y):
    x_phrase = get_phrases(x)
    for phrase in x_phrase:
        phrase = "__".join(phrase)
        phrase_counter[phrase]+=1
```

# Retrieval model using Tf-idf on words

```python
flatten = lambda z: " ".join([flatten(zz) for zz in z]) if isinstance(z,list) and isinstance(z[0],list) else "__".join(z)

X_train,Y_train = read_data('train.loss_fn0')
X_test,Y_test = read_data('test.loss_fn0')
X_valid,Y_valid = read_data('valid.loss_fn0')

h=0
X_train = [flatten(get_phrases(x,height=h)) for x in X_train]
X_test = [flatten(get_phrases(x,height=h)) for x in X_test]
X_valid = [flatten(get_phrases(x,height=h)) for x in X_valid]

# assert len(X) == len(Y)

# X_inv = [flatten(get_phrases(x,height=0)) for x in X]
# assert len(X_inv) == len(Y)

# from sklearn.model_selection import train_test_split
# X_train,X_test,Y_train,Y_test = train_test_split(X_inv,Y,test_size=0.2)

# X_test,X_valid,Y_test,Y_valid = train_test_split(X_test,Y_test,test_size=0.5)

print(len(X_train),len(X_test),len(Y_train),len(Y_test))
print(len(X_test),len(X_valid),len(Y_test),len(Y_valid))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer( max_df = 0.8, stop_words='english')

x_train = vectorizer.fit_transform(X_train)

x_test = vectorizer.transform(X_test)
x_valid = vectorizer.transform(X_valid)
x_test.shape,x_valid.shape,x_train.shape

from sklearn.neighbors import NearestNeighbors
neigh = NearestNeighbors(n_neighbors = 5,n_jobs=-1)
neigh.fit(x_train)
```

```python
module_path = "../src/clai/utils/metric"
if module_path not in sys.path:
    sys.path.append(module_path)
from metric_utils import compute_metric

def compute_scores(Y, Y_hat,nbr_dis):
    scores = []
    i=0
    for nb,true in zip(nbr,Y):
        dis = nbr_dis[i]
        confidence = np.exp(-dis)
        confidence = confidence/confidence[0]
        pred = Y_hat[i]
        inner_score = []
        gz = False
        for p,c in zip(pred,confidence):
            instance_score = compute_metric(p,c,true)
            inner_score.append(instance_score)
            if instance_score>0:
                gz = True
        inner_score = np.array(inner_score)
        if gz:
            scores.append(inner_score.max())
        else:
            scores.append(inner_score.mean())
        i+=1
    return scores
```

```python
import numpy as np
nbr_dis,nbr = neigh.kneighbors(x_valid,return_distance = True)
Y_hat = [[Y_train[n] for n in nb] for nb in nbr]
scores = compute_scores(Y_valid,Y_hat,nbr_dis)
scores = np.array(scores)
import seaborn as sns
sns.displot(scores)
scores.mean(),scores.std()
```

```python
import numpy as np
nbr_dis,nbr = neigh.kneighbors(x_test,return_distance = True)
Y_hat = [[Y_train[n] for n in nb] for nb in nbr]
scores = compute_scores(Y_test,Y_hat,nbr_dis)
scores = np.array(scores)
import seaborn as sns
sns.displot(scores)
scores.mean(),scores.std()
```
