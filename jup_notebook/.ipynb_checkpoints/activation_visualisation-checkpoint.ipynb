```python
import pickle
data_path = '../log/activation_logs.pickle'
with open(data_path,'rb') as f:
    log = pickle.load(f)
```

```python
for k,v in log.items():
    print('Name:#'+k+"#")
    print(len(v))
#     print(v[0])
    print('--'*20)
```

```python
import seaborn as sns
def plot(x):
    x = x.cpu().detach().numpy()
    sns.distplot(x)
    print(x.var())
```

```python
for z in log['0.layer_norm']:
    print(z.shape)
```

```python
for z in log['0.layer_norm']:
    print(z.shape)
    plot(z)
```
